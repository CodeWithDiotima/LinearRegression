```python
from sklearn.datasets import make_regression
from matplotlib import pyplot as plt
import numpy as np
```

```python
X, Y = make_regression(n_samples = 1000, n_features = 1, noise = 15)
```

```python
X.shape
```

```python
Y.shape
```

```python
X = X.flatten()
X.shape
```

```python
plt.scatter(X,Y,marker = '+')
plt.xlabel('x')
plt.ylabel('y')
```

```python
b1 = ((X-np.mean(X)).T).dot((Y-np.mean(Y)))/((X-np.mean(X)).T).dot((X-np.mean(X)))
b1
```

```python
b0 = np.mean(Y) - b1*(np.mean(X))
b0
```

```python
#by using formula of linear regression
print("slope: ",b1," intercept: ",b0)
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(X,Y,marker='+')
ax.plot(X,b1*X+b0,color='red')#step 4
plt.show()
```

```python
#gradient descent with least square method
m = 1
c = 0

theta_list = []
cost_list = []

learning_rate = 0.005
no_of_epochs = 2000
n = float(Y.shape[0])

for i in range(no_of_epochs):
    theta_list.append(m)
    yhat = m*X + c
    err = (1/(2*n))*(np.sum(np.power(Y-yhat,2)))
    #print(err.shape)
    #cost = (1/(2*n))*sum(Y-yhat)
    cost_list.append(err)
    dm = (-1/n)*sum(X * (Y-yhat))
    dc = (-1/n)*sum(Y-yhat)
    m = m - learning_rate*dm
    c = c - learning_rate*dc
    print("Epoch: "+str(i)+"  Error: "+str(err)+"  New Slope Value: "+str(m)+"  New Intercept Value: "+str(c))
#print(cost_list)
print(m,c)
fig, ax = plt.subplots(figsize=(20,20))
#ax.set_xticks([theta_list[0],theta_list[200],theta_list[400],theta_list[600],theta_list[800]])
#ax.set_yticks([cost_list[0],cost_list[200],cost_list[400],cost_list[600],cost_list[800]])
#ax.plot([theta_list[0],theta_list[200],theta_list[400],theta_list[600],theta_list[800]],[cost_list[0],cost_list[200],cost_list[400],cost_list[600],cost_list[800]])#step 4
ax.plot(theta_list,cost_list)
plt.xlabel("theta1")
plt.ylabel("cost")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,20))
ax.plot(range(2000),cost_list)
plt.xlabel("iteration")
plt.ylabel("cost")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(X,Y,marker='+')
ax.plot(X,m*X+c,color='red')#step 4
plt.show()
```

```python
#gradient descent with least square method with different learning rate
m = 1
c = 0

theta_list = []

learning_rate = [0.0001,0.0005,0.001,0.005,0.01]

for l in learning_rate:    
    no_of_epochs = 2000
    n = float(Y.shape[0])
    cost_list = []
    for i in range(no_of_epochs):
        theta_list.append(m)
        yhat = m*X + c
        err = (1/(2*n))*(np.sum(np.power(Y-yhat,2)))
        #cost = (1/(2*n))*sum(Y-yhat)
        cost_list.append(err)
        dm = (-1/n)*sum(X * (Y-yhat))
        dc = (-1/n)*sum(Y-yhat)
        m = m - l*dm
        c = c - l*dc
        #print("Epoch: "+str(i)+"  Error: "+str(err)+"  New Slope Value: "+str(m)+"  New Intercept Value: "+str(c))
    #print(cost_list)
    print("learning rate: ",l)
    print("slope: ",m," intercept: ",c)
    fig, ax = plt.subplots(figsize=(20,20))
    ax.plot(range(2000),cost_list)
    plt.xlabel("theta1")
    plt.ylabel("cost")
    plt.show()
    fig, ax = plt.subplots(figsize=(8,8))
    ax.scatter(X,Y,marker='+')
    ax.plot(X,m*X+c,color='red')#step 4
    plt.show()
```
